```python
import jax
jax.config.update("jax_enable_x64", True)
import jax.numpy as jnp
```

```python
num_points = 10
dim = 2
```

```python
L = jnp.ceil(jnp.log(num_points) / jnp.log(2.0)).astype(int).item()
```

```python
# Compute C array
def compute_C_scan(i: int) -> int:
    # Should convert this to a jax.lax.while_loop...
    init_val = (i, 1)

    cond_fun = lambda x: (x[0] & 1) == 1
    def body_fun(x: tuple) -> tuple: 
        return (x[0]>>1, x[1]+1)

    val = jax.lax.while_loop(cond_fun, body_fun, init_val)

    return val[1]
```

```python
C = jax.vmap(compute_C_scan, in_axes=0)(jnp.arange(1, num_points).astype(jnp.uint32))
```

```python
C = jnp.concatenate([jnp.array([1], dtype=jnp.uint32), C])
```

```python
V = jnp.zeros(L + 1, dtype=jnp.uint32)
for i in range(1, L + 1):
    V = V.at[i].set(1 << (32 - i))
```
